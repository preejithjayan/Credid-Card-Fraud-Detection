```python
#inbalanced data
legit=credit[credit.Class==0]
fraud=credit[credit.Class==1]
```

```python
legit.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
legit_sample=legit.sample(n=492)
```

```python
a=pd.concat([legit_sample,fraud],axis=0)
```

```python
a
```

```python
sns.countplot(x=("Class"),data=a)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
y=a["Class"]
```

```python
x=a.drop("Class",axis=1)
```

```python
x
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=2,stratify =y)
```

```python
model = LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
train_predict=model.predict(xtrain)
```

```python
training_accu=accuracy_score(train_predict,ytrain)
training_accu
```

```python
test_predict=model.predict(xtest)
test_accuracy=accuracy_score(test_predict,ytest)
```
